```python

data = []

start = perf_counter()
for i, url in enumerate(urls, start=1):
  r = requests.get(url, headers=random.choice(headers))
  if r.ok:

    soup = BeautifulSoup(r.content, 'lxml')

    try:
      title = soup.find("span", attrs={"id": 'productTitle'}).string.strip().replace(',', '')
    except AttributeError:
      title = "NA"

    try:
      price = soup.find("span", attrs={'id': 'priceblock_ourprice'}).string.strip().replace(',', '.')
    except AttributeError:
      try:
        price = soup.find("span", attrs={'class': 'a-offscreen'}).string.strip().replace(',', '.')
      except:
        price = "NA"
    
    try:
      rating = soup.find("i", attrs={'class': 'a-icon a-icon-star a-star-4-5'}).string.split()[0].replace(',', '.')
 
    except AttributeError:
      try:
        rating = soup.find("span", attrs={'class': 'a-icon-alt'}).string.split()[0].replace(',', '.')
      except:
        rating = "NA"
    
    try:
      img_url = soup.find("img", attrs={'id': 'landingImage'})['src']
    except:
      try:
        img_url = soup.find("img", attrs={'id': 'imgBlkFront'})['src']
      except:
        img_url = "NA"

    
    print(url, '||', img_url, '||', title, '||', price, '||', rating)
    
    data.append({
        'url': url,
        'img_url': img_url,
        'title': title,
        'price': price,
        'rating': rating
    })

    # wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="ppd"]')))

    #Get title, img_url, price, short_description
    # ppd = soup.find('div', '')
    # title = 
    # img_url =
    # price = 
    # short_description = 


  else:
    if r.status_code == 404:
      print(404, url)
    else:
      print("failed", url)
  

  if i % 100 == 0:
    print(f"{i-100}-{i} pages took {perf_counter() - start} seconds.")

print("\n\n---DONE---\n\n")
print(perf_counter() - start, "seconds taken.")
print(len(data), "pages successfully scraped.")
```
